```python
''' 
    MANIGANDAN V
    22-14-08
    MODELLING AND SIMULATION
    ASSIGNMENT 01            '''



# Prob 01

# MULTIPLYING A REVERSE ARRAY PROBLEM





# Assigning the limit size of an array

N=int(input('Enter the size of an array: '))


'''input an integer list seperated by space from user using map function and split function
then, limiting the array size to N'''

a=list(map(int,input('Enter an array of N numbers:\n').split(' ')[:N]))


# Reversing the list using for loop and append function

b=[]
for i in range(len(a)-1,-1,-1):
    b.append(a[i])
    
    
# Defining recursive function to multiply two lists

def m_list(x,y,i=0):
    if i==len(x):
        return[]
    else:
        return[x[i]*y[i]]+m_list(x,y,i+1)
    
    
# using the recursive function to multiply two lists

c=m_list(a,b)


# printing the resultant array separated by space

print(*c,sep=' ')

```

```python
# Prob 02

# SECOND MAX AND SECOND MIN PROBLEM





# input an integer list separated by space using map and split function

a=list(map(int,input('Enter an array:\n').split(' ')))


# Defining function to find second largest number in a given list

def sec_max(x,n):
    m=-9999999
    for i in range(n):
        if x[i]>m:
            m=x[i]
    s_m=-9999999
    for i in range(n):
        if x[i]>s_m and x[i]<m:
            s_m=x[i]
    return s_m


 # Defining function to find second smallest number

def sec_min(y,n):
    l=9999999
    for i in range(n):
        if y[i]<l:
            l=y[i]
    s_n=999999
    for i in range(n):
        if y[i]<s_n and y[i]>l:
            s_n=y[i]
    return s_n


# assigning n as length of an arry

n=len(a)

# assigning s as list and performing sec_max function

s = [sec_max(a,n)]


# performing sec_min function and store it in s1 variable

s1=sec_min(a,n)

# appending s1 in the list s

s.append(s1)


# printing the result seperated by space by using sep operator

print(*s,sep=' ')
```

```python
# prob 03

# MULTIPLES OF 5 PROBLEM




# input an integer array separated by space

a=list(map(int,input('Enter an array:\n').split(' ')))

# check whether an element of a list is multiples of 5 or not

r=[]
for i in range(len(a)):
    if a[i]%5 != 0:


# if it is not a multiple of 5 then it stores in new list r

        r.append(a[i])
      

 # print the resultant list of numbers separated by space

print(*r,sep=' ')
```

```python
# prob 04

# ONE FLIP PROBLEM




# defining a function to find whether one flip give same sequence of digit(0's and 1's)

def oneFlip(str):
    s=0
    for i in range(len(str)):
        if int(str[i])==1:
            s+=int(str[i])
    return (s==(len(str)-1) or s==1)

# input a string which contais only 0's and 1's


N=input('Enter the 0\'s and 1\'s:\n')


# if the oneFlip function return True it perform if operation otherwise else operation

if oneFlip(N):
    print('yes')
else:
    print('no')
```
